```python
#1. Display country, year, sex, age, suicides/100k pop columns.

import pandas as pd 

dataset = pd.read_csv("suicide.csv")

dataset[["country", "year", "sex", "age", "suicides/100k pop"]]
```

```python
# 2. Display the Philippines' suicide record. Display year, sex, age, suicides_n, and suicides/100k pop columns.
df = dataset.set_index("country")

df.loc["Philippines":"Philippines",["year","sex","age","suicides_no","suicides/100k pop"]]
```

```python
#3. Display the Philippines' suicide record in the year 2011. 
#Display year, sex, age, suicides_n, and suicides/100k pop columns.

df = dataset[(dataset.country == "Philippines") & (dataset.year == 2011)]
df[["country", "year", "sex", "age", "suicides_no", "suicides/100k pop"]]
```

```python
# 4. Display the country with the highest suicide per 100k population 
# in 2005.Display year, sex, age, and suicides/100k pop columns.

get_year = dataset[(dataset.year == 2005)]

sorted_column = get_year.sort_values(by = "suicides/100k pop")

get_highest = dataset[dataset["suicides/100k pop"] == max(sorted_column["suicides/100k pop"])]

get_highest[["country", "year", "sex", "age", "suicides/100k pop"]]
```

```python
# 5. Display the total suicide cases (suicides_n) and suicides/100k pop per year. 
# Sort in descending order based on suicides_n.

get_sum = dataset.groupby(by = "year")[["suicides_no", "suicides/100k pop"]].sum()

get_sum = get_sum.sort_values(by="suicides_no", ascending = False)

get_sum
```

```python
# 6. Display the total suicide cases (suicides_n) and suicides/100k pop per gender. 
# Sort in descending order based on suicides_n.

get_sum = dataset.groupby(by = "sex")[["suicides_no", "suicides/100k pop"]].sum()

get_sum = get_sum.sort_values(by = "suicides_no", ascending = False)

get_sum
```

```python
# 7. Display the total suicide cases (suicides_n) and suicides/100k pop per age. 
# Sort in descending order based on suicides_n.

get_sum = dataset.groupby(by = "age")[["suicides_no", "suicides/100k pop"]].sum()

get_sum = get_sum.sort_values(by = "suicides_no", ascending = False)

get_sum
```

```python
# 8 Display the total suicide cases (suicides_n) per country and per year. 
# Sort in descending order based on suicides_n

get_sum = dataset.groupby(by = ["country", "year"])[["suicides_no"]].sum()

get_sum = get_sum.sort_values("suicides_no", ascending = False)

get_sum
```

```python
# 9. Display the total 
# suicide cases in the Philippines based on year, gender, and age.

df = dataset[(dataset.country == "Philippines")]

get_sum = df.groupby([ "year", "sex", "age"])[["suicides_no"]].sum()

get_sum = get_sum.sort_values("suicides_no", ascending = False)

get_sum
```

```python
# 10. Display the year and number of suicide cases that the Philippines 
# has the highest number of combined suicide cases.

df = dataset[(dataset.country == "Philippines")]

get_sum = df.groupby(["year"])[["suicides_no"]].sum()

max_year = get_sum.idxmax()["suicides_no"]
max_cases = get_sum.max()["suicides_no"]

print(f"Philippines - Year: {max_year} - HighestCases Of Suicide: {max_cases}")
```
