```python
# AGUIRRE, PAUL VINCENT S.
# 1. Display data with collumn Country, year, total water consumption(billion cubix meters)
# water scarcity, agricultural water use(%), industrial water use and household water use

import pandas as pd
df = pd.read_csv("global_water_consumption.csv")
df[["Country","Year","Total Water Consumption (Billion Cubic Meters)","Water Scarcity Level","Agricultural Water Use (%)","Industrial Water Use (%)","Household Water Use (%)"]]
```

```python
# AGUIRRE, PAUL VINCENT S.
# 2. Display the (5)country with highest the total water consumtion in 2024

df[df["Year"] == 2024].sort_values(by="Total Water Consumption (Billion Cubic Meters)", ascending = False).head(5)
```

```python
# AGUIRRE, PAUL VINCENT S.
# 3. Display the (1)country with the lowest total water consumption in 2024 
df[(df.Year == 2024)].sort_values(by = "Total Water Consumption (Billion Cubic Meters)", ascending = True).head(1)
```

```python
# AGUIRRE, PAUL VINCENT S.
# 4. What is the average per capita water use (Liters per Day) for each country across all years?
capita = df.groupby(by = "Country")[["Per Capita Water Use (Liters per Day)"]].mean().sort_values(by = "Per Capita Water Use (Liters per Day)",ascending=False)
capita
```

```python
# AGUIRRE, PAUL VINCENT S.
# 5. Display the global total water consumption trend over the years (using the average per year)

global_trend = df.groupby("Year")[["Total Water Consumption (Billion Cubic Meters)"]].mean().sort_values("Year")
global_trend
```

```python
# AGUIRRE, PAUL VINCENT S.
# 1. Using HVPLOT display which countries had the highest total water consumption in 2024.

import hvplot.pandas

df_2024 = df[df["Year"] == 2024].sort_values(by="Total Water Consumption (Billion Cubic Meters)", ascending=False).head(5)

df_2024_top.hvplot.bar(x="Country", y="Total Water Consumption (Billion Cubic Meters)", title="Top Countries by Water Consumption (2024)", rot=45, color="yellow")
```

```python
# AGUIRRE, PAUL VINCENT S.
# 2. Using HVPLOT display the global average total water consumption change over the years.

global_trend = (df.groupby("Year")[["Total Water Consumption (Billion Cubic Meters)"]].mean().sort_values("Year"))
global_trend.hvplot.line(x="Year", y="Total Water Consumption (Billion Cubic Meters)", title="Global Average Water Consumption Over the Years", color = "orange")
```

```python
# AGUIRRE, PAUL VINCENT S.
# 3. Using HVPLOT to display the per capita water use and annual rainfall
df.hvplot.scatter(x="Per Capita Water Use (Liters per Day)",y="Rainfall Impact (Annual Precipitation in mm)",title="Per Capita Water Use vs. Annual Rainfall",color="blue")
```

```python
# AGUIRRE, PAUL VINCENT S.
# 4. Using HVPLLOT display the average sectoral water use (agricultural, industrial, household) across countries.
df_sectors = (df.groupby("Country")[["Agricultural Water Use (%)", "Industrial Water Use (%)", "Household Water Use (%)"]].mean())
df_sectors.hvplot.bar(x="Country", y=["Agricultural Water Use (%)", "Industrial Water Use (%)", "Household Water Use (%)"], stacked=True, title="Average Sectoral Water Use by Country",rot=45)
```

```python
# AGUIRRE, PAUL VINCENT S.
# 5. Using HVPLOT display the groundwater depletion rate by country.

df.hvplot.box(y="Groundwater Depletion Rate (%)",by="Country",title="Groundwater Depletion by Country", color = "orange", rot = 45)
```
